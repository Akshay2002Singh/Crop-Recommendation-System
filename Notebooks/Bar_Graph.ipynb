```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

ml_accuracies = {
    "LR": 0.94,
    "KNN": 0.97,
    "DT": 0.98,
    "RF": 0.99,
    "NB": 0.99,
    "SVM": 0.95,
    "Bagging": 0.98,
    "AdaBoost": 0.14,
    "Gradient Boosting": 0.98,
    "CatBoost": 0.98,
}

# Convert dictionary to DataFrame for plotting
algorithm_names = list(ml_accuracies.keys())
accuracy_values = list(ml_accuracies.values())
data = {"Algorithm": algorithm_names, "Accuracy": accuracy_values}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x="Algorithm", y="Accuracy", data=df, palette="viridis")
plt.title("Accuracy of Machine Learning Models")
plt.xlabel("Algorithm")
plt.ylabel("Accuracy")
plt.xticks()
plt.tight_layout()
plt.show()
```
